#### 2. Chuyển thành chữ thường

````python
# Converts string into lower case - Chuyển toàn bộ chuỗi thành chữ thường
#Case fold là gì?
'''
    Từ "case fold" trong tiếng Anh có nghĩa là "gấp chữ hoa thành chữ thường". Trong tiếng Việt, chúng ta có thể phân tích ý nghĩa của từ này như sau:

    * "Case" là "vị trí" hoặc "trạng thái". Trong trường hợp này, "case" chỉ vị trí của chữ cái trong bảng chữ cái, bao gồm chữ hoa và chữ thường.
    * "Fold" là "gấp".

    Như vậy, "case fold" có nghĩa là "gấp vị trí của chữ cái trong bảng chữ cái". Điều này có nghĩa là chuyển chữ hoa thành chữ thường.

    Để có thể liên tưởng từ "case fold" có nghĩa là viết thường, chúng ta có thể dựa vào một số điểm sau:

    * Chữ hoa thường được viết bằng cách gấp chữ cái thường lên.
    * Chữ thường thường được viết bằng cách không gấp chữ cái hoa lên.

    Ví dụ:

    * Chữ "A" hoa được viết bằng cách gấp chữ "a" thường lên.
    * Chữ "a" thường được viết bằng cách không gấp chữ "A" hoa lên.

    Như vậy, từ "case fold" có thể được hiểu là "gấp chữ hoa thành chữ thường". Đây là một cách diễn đạt ngắn gọn và súc tích để thể hiện ý nghĩa của từ này.
'''
#Điểm khác nhau giữa hàm casefold() và hàm lower()?
'''
    Hàm casefold() sử dụng bảng mã Unicode để chuyển đổi chữ hoa thành chữ thường. Điều này có nghĩa là hàm casefold() sẽ chuyển đổi tất cả các chữ hoa thành chữ thường, bất kể ngôn ngữ của chuỗi.
    Hàm lower() sử dụng bảng mã ASCII để chuyển đổi chữ hoa thành chữ thường. Điều này có nghĩa là hàm lower() sẽ chỉ chuyển đổi các chữ hoa trong bảng mã ASCII thành chữ thường.
    -> Hàm casefold() phù hợp cho cả so sánh chuỗi phân biệt chữ hoa chữ thường và không phân biệt chữ hoa chữ thường.
    -> Hàm lower() chỉ phù hợp để so sánh chuỗi không phân biệt chữ hoa chữ thường
    --> lower() có tốc độ xử lý nhanh hơn, hiệu suất cao hơn trong khi casefold() chậm hơn vì nó còn phải xử lý các ký tự không phải ASCII hay các ngôn ngữ khác.
'''

string_lower_case = "Here is a SPECIAL STRING to compare Casefold method and Lower method. Now We Have A Special Character In Germany: ß"
print(string_lower_case.casefold())
````

````python
# Converts a string into lower case - Chuyển chuỗi thành chữ thường
print(string_lower_case.lower())
````

````python
# Returns True if all characters in the string are lower case - Trả về True nếu tất cả các ký tự trong chuỗi đều là chữ thường

stringM = "Test String in Python"
print(stringM.islower())
print("test string".islower())
````

````python
# Searches the string for a specified value and returns the position of where it was found - Tìm kiếm chuỗi cho một giá trị được chỉ định và trả về vị trí của nơi nó được tìm thấy
#parameters:
'''
    sub (string) - required, the value to search for
    start (int) - optional, starting position of the search, default is 0
    end (int) - optional, ending position of the search, default is the length of the string
'''

stringM = "Test String in Python"
print(stringM)
print(stringM.find("Str"))
print(stringM.find("Str", 10))
print(stringM.find("in"))
print(stringM.find("in", 10))
````

````python
# Searches the string for a specified value and returns the position of where it was found - Tìm kiếm chuỗi cho một giá trị được chỉ định và trả về vị trí của nơi nó được tìm thấy
#parameters:
'''
    sub (string) - required, the value to search for
    start (int) - optional, starting position of the search, default is 0
    end (int) - optional, ending position of the search, default is the length of the string
'''

stringM = "Test String in Python"
print(stringM.index("in"))
print(stringM.index("in", 5))
print(stringM.index("in", 10))
# if substring not found, it will raise an exception (or error)
# print(stringM.index("abc"))
````

````python
# Creates a translation table to be used in translations - Tạo bảng ánh xạ ký tự, tức là chỉ định danh sách các ký tự cần được thay thế hoặc xóa trong chuỗi.
#Bảng ánh xạ này sau đó được sử dụng với phương thức translate() để thực hiện việc thay thế ký tự theo bảng ánh xạ.
#parameters:
'''
    x (string) - required, a string or a dictionary specifying the characters that should be replaced
        Nếu chỉ có một tham số được chỉ định, thì x phải là một từ điển mô tả cách thực hiện việc thay thế.
        Nếu có hai hoặc nhiều tham số được chỉ định, thì x phải là một chuỗi chỉ định các ký tự bạn muốn thay thế.
    y (string) - optional, a string specifying characters that should be replaced with the characters described in the x parameter. The length of the y string must be equal to the length of the x string
    z (string) - optional, a string specifying characters that should be removed from the string
'''
vowel_dict = {"a": "1", "e": "2", "i": "3", "o": "4", "u": "5"}
x = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
y = "abcdefghijklmnopqrstuvwxyz"
z2 = "abc"
z1 = " "
# Tạo bảng ánh xạ ký tự
translation_table1 = str.maketrans(vowel_dict) # Bằng cách sử dụng từ điển
translation_table2 = str.maketrans(x, y) # Bằng cách sử dụng hai chuỗi
translation_table3 = str.maketrans(x, y, z1) # Bằng cách sử dụng ba chuỗi
translation_table4 = str.maketrans(x, y, z2) # Bằng cách sử dụng ba chuỗi

print(translation_table1)
print(translation_table2)
print(translation_table3)
print(translation_table4)
````

````python
# Returns a translated string - Trả về chuỗi được dịch
#parameters:
'''
    table (string) - required, a translation table containing the mapping between two characters
'''
stringMT_T1 = "This is a test string"
stringMT_T2 = "This is a test string"
stringMT_T3 = "This is a test string"
stringMT_T4 = "This is a test string"

print(stringMT_T1.translate(translation_table1))
print(stringMT_T2.translate(translation_table2))
print(stringMT_T3.translate(translation_table3))
print(stringMT_T4.translate(translation_table4))
````

````python
# Returns a left trim version of the string - Trả về phiên bản cắt bên trái của chuỗi
#parameters:
'''
    chars (string) - optional, a string specifying the set of characters to be removed. If the set of characters is not specified, whitespace will be removed
'''
stringStrip1 = "   Test String   "
stringStrip2 = "***Test String***"
print(stringStrip1.lstrip())
print(stringStrip2.lstrip("*"))
````

````python
# Returns a trimmed version of the string - Xóa khoảng trắng ở đầu và cuối chuỗi và trả về chuỗi mới
#parameters:
'''
    chars (string) - optional, a string specifying the set of characters to be removed. If the set of characters is not specified, whitespace will be removed
'''
stringStrip1 = "   Test String   "
stringStrip2 = "***Test String***"
print(stringStrip1.strip())
print(stringStrip2.strip("*"))
````

##### 1.2. Tách chuỗi thành list

###### a. Split

````python
# Splits the string at the specified separator and returns a list - Tách chuỗi tại dấu phân cách được chỉ định và trả về một danh sách
#parameters:
'''
    separator (string) - optional, specifies the separator to use when splitting the string. Default is any whitespace
    maxsplit (int) - optional, specifies how many splits to do. Default is -1, which means "all occurrences"
'''
stringS = "Is This a test string *** Yes, it is *** Test String in Python ***"
print(stringS.split())
print(stringS.split(" "))
print(stringS.split("*"))
print(stringS.split("*", 2))
print(stringS.split("*", 5))
````

````python
# Splits the string at line breaks and returns a list - Tách chuỗi tại các dấu ngắt dòng và trả về một danh sách
#parameters:
'''
    keepends (bool) - optional, specifies if the line breaks should be included (True), or not (False). Default is False
'''
stringSL = "Is This a test string\nYes, it is\nTest String in Python"
print(stringSL.splitlines())
print(stringSL.splitlines(True))
````

````python
# Splits the string at the specified separator (from the right) and returns a list - Tách chuỗi tại dấu phân cách được chỉ định (từ bên phải) và trả về một danh sách
#parameters:
'''
    separator (string) - optional, specifies the separator to use when splitting the string. Default is any whitespace
    maxsplit (int) - optional, specifies how many splits to do. Default is -1, which means "all occurrences"
'''
print(stringS.rsplit())
print(stringS.rsplit(" "))
print(stringS.rsplit("*"))
print(stringS.rsplit("*", 2))
print(stringS.rsplit("*", 5))
````

###### b. Partition

````python
# Returns a tuple where the string is parted into three parts - Chia chuỗi thành ba phần và trả về một bộ ba dựa trên một ký tự hoặc chuỗi con
#parameters:
'''
    separator (string) - required, specifies the separator to use when splitting the string.
'''
#return:
'''
    A tuple where the string is parted into three parts:
            Phần trước ký tự hoặc chuỗi con sep.
            Ký tự hoặc chuỗi con sep.
            Phần sau ký tự hoặc chuỗi con sep.
'''
stringP = "This is a test string"
print(stringP.partition("test"))
print(stringP.partition("is"))
print(stringP.partition("a"))
print(stringP.partition(" "))
````

````python
#parameters:
'''
    separator (string) - required, specifies the separator to use when splitting the string.
'''
stringRP = "This is a test string"
print(stringRP.rpartition("is"))
print(stringRP.rpartition(" "))
print(stringRP.rpartition("i"))
````

## III. Tuỳ chỉnh hàm sorted() - Sort theo bất kỳ quy tắc nào

Ở mục này, bạn sẽ được học cách khắc phục tất cả những vấn đề ở mục II

Sau khi học, bạn có thể sort theo bất kỳ quy tắc nào, bao gồm sort không phân biệt chữ hoa, chữ thường, sort dict vẫn giữ nguyên value, sort giữa chữ và số và mọi vấn đề khác, chỉ cần bạn nắm vững phần này

Trước khi học, hãy xuống phần đọc hiểu để đọc về hàm lambda để chuẩn bị nền tảng thật tốt trước khi bắt đầu học custom sorted

Trong hàm sorted, ngoài tham số iterator, reverse mà ta đã được học, còn có 1 tham số nữa là key. Bạn có thể hiểu key như 1 quy tắc mà bạn mong muốn nó sắp xếp theo.

-> Cách hoạt động: Khi bạn gán key bằng 1 tên hàm, nó sẽ áp dụng hàm đó cho tất cả các phần tử trong iterator sorted sẽ sắp xếp dựa trên iterator mới
- Ví dụ: bạn có hàm sau: hàm nhận vào 1 số và thực hiện cộng 1 cho số đó
```python
def inverse(x):
    return 1/x
```
- Đồng thời, bạn có 1 lst = [1, 3, 4, 2, 10, 8, 9, 6, 5, 7]
- -> khi bạn gọi hàm sorted(lst, reverse = True, key = inverse), điều đó có nghĩa là bạn sẽ thực hiện lấy nghịch đảo cho lần lượt các phần tử trong iterator và sắp xếp giảm dần iterator mới
- Nghĩa là, bạn sẽ sắp xếp list [1, 1/3, 1/4, 1/2, 1/10, 1/8, 1/9, 1/6, 1/5, 1/7]
- Kết quả nhận về không phải là list [1, 1/3, 1/4, 1/2, 1/10, 1/8, 1/9, 1/6, 1/5, 1/7] đã được sắp xếp, mà là list ban đầu đã được sắp xếp theo quy tắc lấy nghịch đảo

````python
def inverse(x):
    return 1/x

lst = [1, 3, 4, 2, 10, 8, 9, 6, 5, 7]
# khi bạn reverse = True, sẽ sắp xếp theo thứ tự giảm dần
# khi bạn áp quy tắc nghịch đảo (tính chất của nghịch đảo là số ban đầu càng lớn thì nghịch đảo sẽ càng bé và ngược lại)
# -> 1/1 là lớn nhất, còn 1/10 là bé nhất
print(sorted(lst, reverse=True, key=inverse))
````

### 1. Sort theo các quy tắc có sẵn

Các quy tắc hay còn gọi là các hàm có sẵn: bao gồm:

````python
# sort không phân biệt chữ hoa - thường -> bạn muốn xem a cũng như A
# bạn có thể sort theo quy tắc in hoa toàn bộ chuỗi hoặc in thường toàn bộ chuỗi
# -> khi đó a sẽ trở thành A (nếu dùng in hoa) hoặc A thành a (nếu dùng in thường)
# Vậy việc sort sẽ không phân biệt hoa thường
names = ['An', 'Phát', 'Nam', 'an', 'phát', 'nam']
print('sort bình thường', sorted(names))
# vì upper là một method của đối tượng string chứ không phải 1 hàm, bạn có thể lấy ra method bằng cách thông qua đối tượng string
# str.upper()
print('sort bằng tuy tắc in hoa toàn chuỗi', sorted(names, key=str.upper))
print('sort bằng tuy tắc in thường toàn chuỗi', sorted(names, key=str.lower))
````

````python
# khi bạn muốn sort giữa chữ và số, hãy sort theo quy tắc ép số thành chuỗi
# vì 1 số chắc chắn sẽ có thể chuyển thành chuỗi, còn 1 chuỗi nếu không chứa nội dung là số sẽ không thể ép nên ta chọn ép số thành chuỗi
# lúc này, các số sẽ được xem là những ký số (digit), vd 2 -> ký tự 2, 1 -> ký tự 1...
lst = [2, 1, 10, 11, 9, 7, 6, 8, 5, 3, 4, 12, 13, 15, 14, 16, 17, 'a', 'b', 'c', 'dbmxn']
sorted_lst = sorted(lst, key=str)
print(sorted_lst) # -> in thành công, chương trình không hề báo lỗi
````

#### Sort với dict

````python
# Trước tiên, hãy ôn lại về method lấy ra các items

phat_infor = {
    'tên': 'Phát'
    , 'tuổi': 19
    , 'is_male': True
    , 'ngành': 'CNTT'
    , 'sở_thích': ('đi du dịch', 'tập gym', 'chơi game')
    , 'crush': None
}
print(phat_infor.items())
````

````python
# để sort 1 dict, ta có thể sắp xếp theo quy tắc sắp xếp theo key và sắp xếp theo value
phat_infor = {
    'tên': 'Phát'
    , 'tuổi': 19
    , 'is_male': True
    , 'ngành': 'CNTT'
    , 'sở_thích': ('đi du dịch', 'tập gym', 'chơi game')
    , 'crush': None
}
# truyền vào phat_infor.items() để lấy được các cặp tuple ( key, value ). Nếu không, phat_infor chỉ là 1 list các key
# với mỗi phần tử là 1 tuple, gồm 2 vị trí là key và value -> vị trí 0 là key, vị trí 1 là value
# -> muốn sắp xếp theo key thì dùng lambda x: x[0] để lấy ra giá trị của index 0 trong mỗi phần tử (tuple)
sorted_phat_infor_key = sorted(phat_infor.items(), key=lambda x: x[0])
print(sorted_phat_infor_key)

# -> muốn sắp xếp theo value thì dùng lambda x: x[1] để lấy ra giá trị của index 1 trong mỗi phần tử (tuple)
# bạn thấy value mang nhiều kiểu dữ liệu khác nhau -> hãy ép nó về string để tránh báo lỗi nhé
sorted_phat_infor_value = sorted(phat_infor.items(), key=lambda x: str(x[1]))
print(sorted_phat_infor_value)
````

### 2. Sort theo 1 quy tắc tự định nghĩa

sort theo quy tắc hay hàm tự định nghĩa là tự viết ra 1 hàm để làm quy tắc sort khi nhận thấy python không hỗ trợ quy tắc mà ta mong muốn

````python
# sort theo quy tắc nghịch đảo

def inverse(x):
    return 1/x

lst = [1, 3, 4, 2, 10, 8, 9, 6, 5, 7]
print(sorted(lst, key=inverse))
````

````python
# giả sử 1 list với các phần tử là list, và ta muốn sort theo trung bình của mỗi list con
lst = [[1, 3, 5], [2, 4, 6], [3, 5, 7], [1, 4, 2]]

def average_list(lst):
    return sum(lst) / len(lst)

sorted_lst = sorted(lst, key=average_list)
print(sorted_lst)
````
