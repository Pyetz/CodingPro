````python
# phần 1: header - là nơi khai bao các module (thư viện cần thiết)
import math
import numpy as np
import random as r

# phần 2: các hàm cần thiết
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def is_prime_v2(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def is_prime_v3(n):
    if n < 2:
        return False
    for i in range(2, n):
        if i * i > n:
            break
        if n % i == 0:
            return False
    return True

# phần 3: Hàm main()
def main():
    n = 1000000
    count = 0
    for i in range(n):
        if is_prime_v3(i):
            count += 1
    print(count)

# phần 4: gọi hàm main()
if __name__ == "__main__":
    main()
````

#### Bài 1

Viết chương trình kiểm tra các một số có phải là số dương hay không, nếu có thì in ra 'YES', ngược lại in ra 'NO'
- Input:
    - dòng 1 là số lượng bộ test T
    - sau đó là T dòng với mỗi dòng là 1 số cần kiểm tra
    - ... các ràng buộc về số ...
- Output:
    - kết quả YES/NO trên mỗi dòng tương ứng với input
- Ví dụ:

````python
input:              |   Output:
5                   |   YES
10                  |   YES
2                   |   NO
-10                 |   NO
0                   |   NO
-1                  |
````

````python
def is_positive(num):
    return num > 0

def get_input():
    tess = int(input("Nhập số lượng bộ tess: "))
    test_cases = []
    for _ in range(tess):
        n = int(input("Nhập số nguyên n: "))
        test_cases.append(n)

    return test_cases

def main():
    test_cases = get_input()

    for n in test_cases:
        if is_positive(n):
            print("YES")
        else:
            print("NO")

if __name__ == "__main__":
    main()
````

#### Bài 2

Viết chương trình kiểm tra tổng các số trên 1 dòng có lớn hơn không hay không
- Input:
    - dòng 1 là số lượng bộ test T
    - sau đó là T dòng với mỗi dòng là 1 dãy các số cần kiểm tra
    - ... các ràng buộc về số ...
- Output:
    - kết quả YES/NO trên mỗi dòng tương ứng với input
- Ví dụ:

````python
input:              |   Output:
3                   |   NO
10 2 4 8 6 4 1 -100 |   YES
2 3 4 -10 5 7       |   NO
-10 -100 500 -400   |
````

````python
def is_positive(line):
    return sum(line) > 0

def get_input():
    tess = int(input("Nhập số lượng bộ tess: "))
    test_cases = []
    for _ in range(tess):
        line = map(int, input().split())
        test_cases.append(line)

    return test_cases

def main():
    test_cases = get_input()

    for lst in test_cases:
        if is_positive(lst):
            print("YES")
        else:
            print("NO")

if __name__ == "__main__":
    main()
````

#### Bài 3

Viết chương trình nhập vào 1 ma trận nxm (n hàng m cột) và in ra ma trận đó
- Input:
    - dòng đầu nhập vào 2 số n và m
    - n dòng tiếp theo là các phần tử trên 1 hàng ngăn cách bởi dấu phẩy
    - ... các ràng buộc về số ...
- Output:
    - in ra các ma trận
- Ví dụ:

````python
input:              |   Output:
4 5                 |   1 2 4 2 1
1 2 4 2 1           |   3 3 2 5 3
3 3 2 5 3           |   -1 -2 -10 -5 -100
-1 -2 -10 -5 -100   |   20 30 1 2 3
20 30 1 2 3         |
````

````python
def input_matrix():
    n, m = map(int, input().split())
    matrix = []
    for _ in range(n):
        row = list(map(int, input().split()))
        matrix.append(row)

    return matrix

def main():
    matrix = input_matrix()

    for line in matrix:
        ' '.join(map(str, line))

if __name__ == "__main__":
    main()
````

#### Bài 4

Viết chương trình nhập vào 1 ma trận nxm (n hàng m cột) và in ra ma trận đó
- Input:
    - dòng 1 là số lượng bộ test T
    - sau đó là T lần, mỗi lần
        - dòng đầu nhập vào 2 số n và m
        - n dòng tiếp theo là các phần tử trên 1 hàng ngăn cách bởi dấu phẩy
    - ... các ràng buộc về số ...
- Output:
    - in ra các ma trận
- Ví dụ:

````python
input:              |   Output:
2                   |   5 3
2 2                 |   2 4
5 3                 |   1 2 4 2 1
2 4                 |   3 3 2 5 3
4 5                 |   -1 -2 -10 -5 -100
1 2 4 2 1           |   20 30 1 2 3
3 3 2 5 3           |
-1 -2 -10 -5 -100   |
20 30 1 2 3         |
````

````python
def input_matrix():
    tess = int(input("Nhập số lượng ma trận: "))
    test_cases = []
    for _ in range(tess):
        n, m = map(int, input().split())
        matrix = []
        for _ in range(n):
            row = list(map(int, input().split()))
            matrix.append(row)
        test_cases.append(matrix)

    return test_cases

def main():
    test_cases = input_matrix()

    for matrix in test_cases:
        for row in matrix:
            print(row)

if __name__ == "__main__":
    main()
````
