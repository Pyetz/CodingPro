## BT1: Kiểm tra chia hết/không chia hết

Viết hàm kiểm tra một số có phải là số vừa chia hết cho 3 vừa chia hết cho 5 nhưng không chia hết cho 7 không, đúng trả về True, sai trả về False

```python
def check_divisibility(number):
    if number % 3 == 0 and number % 5 == 0 and number % 7 != 0:
        return True
    else:
        return False
```

```python
# test
test_cases = 4
true = 0
if check_divisibility(123456789) == False:
    true += 1
if check_divisibility(1234567890) == True:
    true += 1
if check_divisibility(105) == False:
    true += 1
if check_divisibility(123456789012) == False:
    true += 1

print(f"{true}/{test_cases} test cases passed")
```
