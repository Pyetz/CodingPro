### 4.3. Những lưu ý về string

#### a. Cách nhận biết và khai báo 1 string

String (chuỗi) là tất cả mọi thứ nằm bên trong cặp ngoặc (có thể bao gồm chữ, số, và cả các ký tự đặc biệt, ... Chỉ cần nằm trong 1 trong 3 cặp ngoặc sau:
1. Cặp ngoặc đơn: ```'string'```
2. Cặp ngoặc đôi: ```"string"```
3. Cặp 3 ngoặc đơn (thường dùng để chứa 1 string có nhiều dòng): ```'''string'''```
4. Cặp 3 ngoặc đôi (thường dùng để chứa 1 string có nhiều dòng): ```"""string"""```

````python
# 4 dòng code bên dưới đều như nhau, bạn muốn viết kiểu nào cũng được
# khuyến khích dùng cách 1 hoặc 2, còn cách 3, 4 dùng cho string có nhiều dòng
str1 = 'Hello, World!'
str2 = "Hello, World!"
str3 = '''Hello, World!'''
str4 = """Hello, World!"""

# Dù viết cách nào thì nó vẫn là string. Bạn có thể kiểm tra bằng hàm type()
print(type(str1))
print(type(str2))
print(type(str3))
print(type(str4))
````

````python
# string có nhiều dòng

string_multiline1 = """This is a
multiline string"""

string_multiline2 = '''This is a
multiline string'''

# Thử in ra bằng hàm print nhé, bạn thấy nó gồm nhiều dòng.
# string nhiều dòng thì kiểu dữ liệu vẫn là string. Kiểm tra bằng hàm type()
print('string_multiline1: ', string_multiline1)
print(type(string_multiline1))

print() # print() sẽ thể in ra 1 dòng trống cho dễ nhìn

print('string_multiline2: ', string_multiline2)
print(type(string_multiline2))
````

## 5. Ép kiểu dữ liệu (Casting)

Ép kiểu dữ liệu là chuyển đổi kiểu dữ liệu 1 số/chuỗi

### 5.1. Chuyển 1 số nguyên hoặc 1 số thực sang 1 string

Bạn có thể chuyển bất cứ 1 số nguyên hay 1 số thực nào sang string bằng hàm str() mà không gặp bất cứ lỗi gì

````python
# giả sử có biến num1 là số nguyên, num2 là số thực
num1 = 25
num2 = 25.07

# kiểm tra kiểu dữ liệu của 2 biến trên
print(type(num1))
print(type(num2))
````

````python
# Bạn có thể chuyển nó sang 1 string bằng hàm str()

str_num1 = str(num1)
str_num2 = str(num2)

# thử in ra bằng hàm print nhé
print(str_num1)
print(str_num2)

# kiểm tra kiểu dữ liệu của 2 biến sau khi chuyển
print(type(str_num1))
print(type(str_num2))
````

### 5.2. Chuyển 1 chuỗi sang 1 số

Để chuyển 1 chuỗi sang số cần đảm bảo nội dung trong chuỗi là 1 số

#### a. Chuyển mỗi chuỗi sang số nguyên

````python
string_convert = '25'
num = int(string_convert)

# thử in ra bằng hàm print nhé
print(num)

# kiểm tra kiểu dữ liệu của biến sau khi chuyển
print(type(num))
````

````python
string_convert = '25.07'
num = int(string_convert)
````

#### a. Chuyển mỗi chuỗi sang số thực bằng hàm float()

````python
string_convert = '25.07'
num = float(string_convert)

# thử in ra bằng hàm print nhé
print(num)

# kiểm tra kiểu dữ liệu của biến sau khi chuyển
print(type(num))
````

## 2. Python có phát hiện lỗi trước khi bạn chạy code của mình không?

Để 1 chương trình được thực thi, cần có 1 trình dịch. CÓ 2 loại trình dịch:
- Trình biên dịch (compiler): như trong C++, khi ấn thực thi 1 đoạn code thì toàn bộ code sẽ được chuyển sang mã máy, nên nếu code có bất kỳ lỗi nào thì chương trình sẽ không chạy được
- Trình thông dịch (interpreter): thay vì chạy 1 lúc toàn bộ code, trình thông dịch sẽ chạy từng dòng lệnh

Python sử dụng trình thông dịch nên chương trình sẽ chạy tới đúng đoạn có lỗi -> Mình có thể tận dụng điều này để biết khúc nào có lỗi.
